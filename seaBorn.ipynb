```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

plt.sca(ax[0])
sns.histplot(x='total_bill', data=df_tips[df_tips['smoker'] == 'No'], bins=25)
## title and axis
plt.title('Distr. of total bill in Dataset', fontsize=14)
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.sca(ax[1])
sns.histplot(x='total_bill', data=df_tips[df_tips['smoker'] == 'Yes'], bins=25)
## title and axis
plt.title('Distr. of total bill in Dataset', fontsize=14)
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.show()
```
