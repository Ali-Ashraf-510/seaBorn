```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df_tips, hue='sex', size='size', style='time', palette='coolwarm')


# بيرسم مخطط انتشار (Scatter Plot) يوضح العلاقة بين total_bill وtip باستخدام بيانات df_tips.

```

```python
sns.relplot(x='total_bill', y='tip', data=df_tips , kind ='scatter',row='time',col='sex' )
#رسم مخططات انتشار متعددة حسب الجنس في أعمدة ووقت اليوم في صفوف
```

```python
sns.relplot(x='total_bill', y='tip', data=df_tips , kind ='scatter',row='time',col='smoker' )
```

```python
sns.boxplot(x='sex',y='tip',data=df_tips,hue='smoker')
# بيرسم Box Plot (رسم صندوقي) لمقارنة توزيع الإكرامية (tip) بناءً على الجنس (sex)، مع التفريق حسب التدخين (smoker).

# 📊 شرح المعلمات:
# x='sex': المحور الأفقي يمثل الجنس.
# y='tip': المحور العمودي يمثل الإكرامية.
# data=df_tips: مصدر البيانات.
# hue='smoker': يلون الصناديق بناءً على التدخين (مدخن/غير مدخن).
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df_tips,size='size') 

# = 

# sns.relplot(x='total_bill',y='tip',data=df_tips,kind='scatter')
```

```python
 # إذا أردت المقارنة في نفس الشكل → استخدم hue.
 # إذا أردت مخططات منفصلة لكل فئة → استخدم col.
```

```python
sns.relplot(x='total_bill',y='tip',data=df_tips,size='size',hue='sex',row='time',kind='scatter') 
```

```python
sns.relplot(x='total_bill',y='tip',data=df_tips,size='size',row='sex',col='time',kind='scatter',hue='day',style='smoker') 
```

```python
sns.histplot(x='tip',data=df_tips,bins=25,hue='smoker',kde=True)

# x='tip': المتغير على المحور الأفقي (قيمة الإكرامية).
# data=df_tips: مصدر البيانات من df_tips.
# bins=25: عدد الأعمدة في المخطط (مقسّم لـ 25 فئة).
# hue='smoker': لون مختلف للمدخنين وغير المدخنين.
# kde=True: يرسم منحنى الكثافة (KDE) فوق المخطط لإظهار الشكل العام للتوزيع.
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

plt.sca(ax[0])
sns.histplot(x='total_bill', data=df_tips[df_tips['smoker'] == 'No'], bins=25)
## title and axis
plt.title('Distr. of total bill in Dataset', fontsize=14)
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.sca(ax[1])
sns.histplot(x='total_bill', data=df_tips[df_tips['smoker'] == 'Yes'], bins=25)
## title and axis
plt.title('Distr. of total bill in Dataset', fontsize=14)
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.show()
```
